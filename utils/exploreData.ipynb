```python
import os
import numpy as np
import tifffile as tiff

def transform_and_save_data(data_dir, output_file):
    """
    Transforms the data from the given directory structure and saves it as a .npz file.
    
    Args:
    - data_dir (str): The directory containing the tif images.
    - output_file (str): The path to save the output .npz file.
    
    Returns:
    None
    """
    # Initialize a list to hold the data
    data_list = []
    
    # Loop over each sub-directory (location/S1_S2)
    print(data_dir)
    
    tif_files = [f for f in sorted(os.listdir(data_dir)) if f.endswith('.tif') and 'slope' not in f and 'elevation' not in f]
    
    # Read the 4 images
    images = [tiff.imread(os.path.join(data_dir, f)) for f in tif_files]

    # # Read the 4 images in the directory
    # images = [tiff.imread(os.path.join(full_path, f)) for f in sorted(os.listdir(full_path)) if f.endswith('.tif') and 'elevation' not in f and 'slope' not in f]

    # Stack the images along a new dimension to get shape [4, 410, 256, 256]
    images_stack = np.stack(images)
    
    print(images_stack.shape)
    # Split the stacked images into dynamic and static parts
    dynamic_images = images_stack[:, :, :, :408]
    static_images = images_stack[:, :, :, 408:]

    # print(dynamic_images.shape)
    # Reshape dynamic images to [4, 17, 24, 256, 256]
    reshaped_dynamic = dynamic_images.reshape(4, 17, 24, 256, 256)
    
    # Repeat static images across the time dimension to match dynamic images
    repeated_static = np.repeat(static_images[:, :, np.newaxis, :, :], 24, axis=2)
    repeated_static = repeated_static.reshape(4, 2, 24, 256, 256)
    print(repeated_static.shape)
    print(reshaped_dynamic.shape)
    # Concatenate dynamic and static images along the band dimension
    combined_images = np.concatenate((reshaped_dynamic, repeated_static), axis=1)
    
    # Move the time dimension to the second position to get [4, 17, 1, 24, 256, 256]
    combined_images = combined_images.transpose(0, 1, 2, 3, 4)
    
    final_images = combined_images[:, :, :, np.newaxis, :, :]

    print('Final', final_images.shape)
    # Save the final data as a .npz file
    final_images1 = final_images[:, :, :12, :, :, :]
    final_images2 = final_images[:, :, 12:, :, :, :]

    np.savez_compressed(output_file[:-5]+'1'+output_file[-5:], data=final_images1)
    np.savez_compressed(output_file[:-5]+'2'+output_file[-5:], data=final_images2)




data_dir = '/workspaces/presto/presto/exampleLoc'
output_file = '/workspaces/presto/presto/exampleLocOut/output_data.npz'
transform_and_save_data(data_dir, output_file)
```

```python
import json
import numpy as np

# Function 1: Load the GeoJSON File
def load_geojson(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

# Function 2: Count Features for Each Biome Name and Each Label
def count_features(data):
    biome_counts = {}
    label_counts = {
        "Bare_Ground": 0,
        "Built_Area": 0,
        "Clouds": 0,
        "Crops": 0,
        "Flooded_Vegetation": 0,
        "Grass": 0,
        "Scrub": 0,
        "Snow_Ice": 0,
        "Trees": 0,
        "Water": 0
    }
    
    for feature in data['features']:
        biome_name = feature['properties']['biome_name']
        if biome_name in biome_counts:
            biome_counts[biome_name] += 1
        else:
            biome_counts[biome_name] = 1
        
        for label in label_counts:
            if feature['properties'][label] > 0:
                label_counts[label] += 1
    
    # Save counts to numpy files
    np.save('biome_counts.npy', biome_counts)
    np.save('label_counts.npy', label_counts)
    
    return biome_counts, label_counts

# Function 3: Save a Subset of Features Based on Specific Conditions
def save_subset_geojson(data, output_file):
    subset_features = []
    countIn = 0
    countOut = 0
    for feature in data['features']:
        if (feature['properties']['Bare_Ground'] > 0 or
            feature['properties']['Crops'] > 0 or
            feature['properties']['Flooded_Vegetation'] > 0 or
            feature['properties']['Grass'] > 0 or
            feature['properties']['Scrub'] > 0 or
            feature['properties']['Snow_Ice'] > 0 or
            feature['properties']['Trees'] > 0):
            subset_features.append(feature)
            countIn +=1
        else:
            countOut +=1
    
    print('Kept subset ', countIn, ' out of total ', countOut)
    
    subset_data = {
        "type": "FeatureCollection",
        "crs": data["crs"],
        "features": subset_features
    }
    
    with open(output_file, 'w') as f:
        json.dump(subset_data, f)

def plot_label_counts(label_counts):
    labels = list(label_counts.keys())
    counts = list(label_counts.values())

    plt.figure(figsize=(10, 6))
    plt.bar(labels, counts, color='skyblue')
    plt.xlabel('Labels')
    plt.ylabel('Counts')
    plt.title('Counts of Each Label')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Example usage
file_path = '/workspaces/presto/data/dynamic_world_samples.geojson'
data = load_geojson(file_path)
biome_counts, label_counts = count_features(data)
save_subset_geojson(data, '/workspaces/presto/data/dynamic_world_subsamples.geojson')
plot_label_counts(label_counts)

print("Biome counts:", biome_counts)
print("Label counts:", label_counts)
```
